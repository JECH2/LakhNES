```python
def get_log(filename):
             
    Log = {}
    Log['train_epoch'] = []
    Log['train_step'] = []
    Log['train_lr'] = []
    Log['train_loss'] = []
    Log['train_ppl'] = []

    Log['val_step'] = []
    Log['val_loss'] = []
    Log['val_ppl'] = []
    with open(filename, "r") as f:
        lines = f.readlines()
        for line in lines:
            line = line.split()
            # epoch
            if (len(line) > 2 and line[1] =='epoch'):
                # epoch / step(batch) / lr / loss / ppl
                Log['train_epoch'].append(float(line[2]))
                Log['train_step'].append(float(line[4]))
                Log['train_lr'].append(float(line[10]))
                Log['train_loss'].append(float(line[16]))
                Log['train_ppl'].append(float(line[19]))
                # print(line[2], line[4], line[10], line[16], line[19])
            # validation
            if (len(line) > 2 and line[1] =='Eval'):
                Log['val_step'].append(float(line[5]))
                Log['val_loss'].append(float(line[12]))
                Log['val_ppl'].append(float(line[16]))
                #print(line[2],line[5], line[12], line[16])

    
    return Log
```

```python
from matplotlib import pyplot as plt

# draw each line in X, Y and scatter dots (x_sample, y_sample)
def draw_plot(X, Y):
    plt.plot(X, Y)
    plt.show()
```

```python
filename1 = "model/pretrained/Lakh400kPretrainOnly/log.txt"
filename2 = "results/-nesmdb/20200917-031518/log.txt"
log1 = get_log(filename1)
log2 = get_log(filename2)

plt.figure(figsize=(20,5))
plt.subplot(121)
plt.plot(log2['train_step'], log2['train_loss'])
plt.plot(log2['val_step'][1:], log2['val_loss'][1:])
plt.xlabel('train_step')
plt.title('Loss of Lakh Train Reproduce')
plt.legend(['train loss', 'val loss'])

plt.subplot(122)
plt.plot(log1['train_step'], log1['train_loss'])
plt.plot(log1['val_step'], log1['val_loss'])
plt.xlabel('train_step')
plt.title('Loss of Lakh400k Pretrain Only')
plt.legend(['train loss', 'val loss'])
plt.show()
plt.savefig('Reproduce_Lakh_loss_20200920.png')
```

```python
plt.figure(figsize=(20,5))
plt.subplot(121)
plt.plot(log2['train_step'], log2['train_ppl'])
plt.plot(log2['val_step'][1:], log2['val_ppl'][1:])
plt.xlabel('train_step')
plt.title('Perplexity of Lakh Train Reproduce')
plt.legend(['train ppl', 'val ppl'])

plt.subplot(122)
plt.plot(log1['train_step'], log1['train_ppl'])
plt.plot(log1['val_step'], log1['val_ppl'])
plt.xlabel('train_step')
plt.title('Perplexity of Lakh400k Pretrain Only')
plt.legend(['train ppl', 'val ppl'])
plt.show()
plt.savefig('Reproduce_Lakh_ppl_20200920.png')
```
